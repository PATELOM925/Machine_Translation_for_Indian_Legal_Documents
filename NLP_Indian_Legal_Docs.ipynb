```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text
```

## N-Grams

```python
def find_legal_terms(input, legalKeys):
    processed_keys = [preprocess(key) for key in legalKeys]
    tokens = word_tokenize(preprocess(input))
    bigrams = list(ngrams(tokens, 2))
    trigrams = list(ngrams(tokens, 3))
    fgrams = list(ngrams(tokens, 4))

    bigram_strings = [' '.join(bigram) for bigram in bigrams]
    trigram_strings = [' '.join(trigram) for trigram in trigrams]
    fgram_strings = [' '.join(fgram) for fgram in fgrams]
    legal_terms = set()     #matching tokens with legal keys

    for token in tokens: #finding a match for single token with legal_term's key
        if token in processed_keys:
            legal_terms.add(token)

    for ngram in bigram_strings + trigram_strings + fgram_strings :
        if ngram in processed_keys: # Finding a match for group of tokens with legal_term key
            legal_terms.add(ngram)

    return legal_terms
```

```python
def simplify(input_text, legal_dict):
    legal_keys = legal_dict.keys()
    matched_legal_terms = find_legal_terms(input_text, legal_keys)
    tokens = word_tokenize(input_text)  #tokenize the input text
    simplified_sentence = []

    # Subsituting matched legal terms with their definitions in bracket
    for token in tokens:
        preprocessed_token = preprocess(token)
        if preprocessed_token in matched_legal_terms:
            simplified_sentence.append(f" '{token}':({legal_dict[preprocessed_token]})")
        else:
            simplified_sentence.append(token)

    return ' '.join(simplified_sentence)
```

##Substituting

```python
def replace(input_text, legal_dict):
    legal_keys = legal_dict.keys()
    matched_legal_terms = find_legal_terms(input_text, legal_keys)
    tokens = word_tokenize(input_text)  #tokenize the input text
    simplified_sentence = []

    # Subsituting matched legal terms with their definitions in bracket
    for token in tokens:
        preprocessed_token = preprocess(token)
        if preprocessed_token in matched_legal_terms:
            simplified_sentence.append(f"({legal_dict[preprocessed_token]})")
        else:
            simplified_sentence.append(token)

    return ' '.join(simplified_sentence)
```

## Paraphrase

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

model_name_p = "t5-base"
tokenizer_p = T5Tokenizer.from_pretrained(model_name_p)
model_p = T5ForConditionalGeneration.from_pretrained(model_name_p)
```

```python
# Function to perform final Paraphrasing
def summarize_text(text):
    input_text = "summarize: " + text
    p_inputs = tokenizer_p.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
    p_summary_ids = model_p.generate(p_inputs, max_length=150, min_length=50, length_penalty=2.0, num_beams=4, early_stopping=True)
    p_summary = tokenizer_p.decode(p_summary_ids[0], skip_special_tokens=True)
    return p_summary
```
